```python
import pandas as pd
import json
```

```python
with open('data/filenames.txt') as f:
    lines = [line.strip() for line in f.readlines()]
lines
```

```python
CID_dict = {}
for file in lines:
    CID = file.split('/')[-1].split('.')[0] + '-' + file.split('/')[-1].split('.')[1]
    CID_dict[CID] = file
```

```python
meta = pd.read_csv('data/ut-zap50k-data/meta-data.csv')
meta.head()
```

```python
meta['filename'] = meta["CID"].apply(lambda x: CID_dict.get(x))
meta.head()
```

```python
closures = ['Lace up', 'Slip-On', 'Zipper', 'Hook and Loop', 'Pull-on']
genders = ['Women', 'Men', 'Girls', 'Boys']
meta = meta[meta['Closure'].isin(closures)]
meta = meta[meta['Gender'].isin(genders)]
```

```python
meta.shape
```

```python
import random
random.seed(2)
```

```python
meta.shape
```

```python
val_test_indices = random.sample(range(meta.shape[0]), 7560)
val_indices = val_test_indices[:3780]
test_indices = val_test_indices[3780:]
```

```python
meta['split'] = 'train'
meta['split'].iloc[val_indices] = 'val'
meta['split'].iloc[test_indices] = 'test'
```

```python
meta.head()
```

```python
category = 'Closure'
print('Train')
print(meta[meta['split'] == 'train'][category].value_counts())
print('Val')
print(meta[meta['split'] == 'val'][category].value_counts())
print('Test')
print(meta[meta['split'] == 'test'][category].value_counts())
```
